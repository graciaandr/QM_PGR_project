```python
import pandas as pd
filepath = "/Users/graciaandriamiadana/Documents/Research Culture PGR project/data/PGR_disciplines.csv"
df = pd.read_csv(filepath) 
```

```python
from itertools import combinations

# Assuming you have loaded your data into a DataFrame named df

# Extract unique combinations of disciplines
unique_combinations = set()
for disciplines_list in df['Disciplines']:
    print(disciplines_list)
    unique_combinations.update(combinations(disciplines_list, 2))  # Get combinations of length 2

# If the length of unique combinations is greater than 1, it indicates interdisciplinary research
if len(unique_combinations) > 1:
    print("The research is interdisciplinary.")
else:
    print("The research is not interdisciplinary.")


len(unique_combinations)
```
